```
import pandas as pd
import matplotlib.pyplot as plt # Para visualizar datos <----- Le agregué pyplot

from sklearn.model_selection import train_test_split #Modelo de selección tomado de la biblioteca
#Recuerda entrenar y poner a pruebas nuestros modelos

from sklearn.linear_model import LinearRegression #Ubicamos los modelos lineales y escogemos Regresióón Lineal

```

```
dataset = pd.read_csv('salarios.csv') #Conjunto de datos o dataset
dataset.head()
```

```
#Cuántos ejemplos, casos o datos tenemos
dataset.shape
```

```
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,1].values #La parte que predecirá 
#ejecutamos para cargar valores
```

```
X_train, X_test, Y_train, Y_test = train_test_split(x,y,test_size=.2, random_state=0) 
#Es parte deskilearn
#Lo usaremos para dividir información

#####Test_size 
#para indicar como usaremos los datos o vaya la cantidad usada
#Recuerda 70% de la data (aproximadamente) -> Para entrenamientos
#NOTA: Recuerda que es 1- la cantidad no usada de datos

#####Random_State
#-> Para modificar los datos -> Por el momento no modificaremos ningun dato
```

```
X_train #Informacióón que usaremos para predecir
```

```
#Creación del modelo 

regressor = LinearRegression()
regressor.fit(X_train,Y_train) #Entrenamos el modelo (fit :v)

#Recapitulemos
# Datos -> Dividir datos -> Entrenar 
```

```
viz_train = plt #Utilizamos matplot para graficar
viz_train.scatter(X_train,Y_train, color ='blue')#Pasamos los valores a nuestra gráfica
viz_train.plot(X_train,regressor.predict(X_train),color ='black')
viz_train.title('Salario vs Experiencia')
viz_train.xlabel('Experiencia')
viz_train.ylabel('Salario')
viz_train.show()
```

Y pues tenemos nuestro primer modelo, datos alejados son los que no ha encotrado una relación adecuada.

```
#¿Cóómo sabemos que estáá bien?
#Por lo tanto datos de prueba
viz_train = plt #Utilizamos matplot para graficar
viz_train.scatter(X_test,Y_test, color ='red')#Pasamos los valores a nuestra gráfica
viz_train.plot(X_train,regressor.predict(X_train),color ='black')
viz_train.title('Salario vs Experiencia')
viz_train.xlabel('Experiencia')
viz_train.ylabel('Salario')
viz_train.show()
```

Tenemos pocos datos para validar nuestros datos :( -> Podemos agregar máás datos y así validar que tanto aprenderemos

```
regressor.score(X_test,Y_test)
```

Matemáticamente conocido como Coeficiente de correlacióón, es una forma de validar que tan preciso es en este momento nuestro modelo al prónosticar datos



```
#Trabajo de Clase hacer una regresión lineal multiple con los mismos datos
import numpy as np #Para valores aleatorios

dataset = pd.read_csv('salarios.csv')
paises = ['VZ','MX','CO','BR','AR','US'] #Ponemos los valores de los paises

#Extraemos valores de forma aleatorios y lo hacemos mediante una iteración
new_column = [paises[np.random.randint(0,len(paises))] for i in range(30) ]

#Asignamos valores
dataset['Pais'] = new_column

dataset['NormPaises'] = pd.factorize(dataset['Pais'])[0]

#Asignamos conjuntos -> En el caso de X tenemos dos valores
x = dataset[['Aexperiencia','NormPaises']]
y = dataset['Salario']

#Literalmente copiamos lo anterior
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size = 0.6, random_state=0)#Escogemos esta cantidad de data

#Entrenamos y vemos el score del modelo
regressor.fit(X_train,y_train)
regressor.score(X_test,y_test)
```

```
#Parte de gráficas 

from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from matplotlib import style

fig = plt.figure(dpi = 150)
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_train['Aexperiencia'],X_train['NormPaises'],y_train,c='b',marker='s') #Cuadrados son valores de entrenamiento
ax.scatter(X_train['Aexperiencia'],X_train['NormPaises'],regressor.predict(X_train),c='r',marker='o') #Puntos son valores predichos
ax.set_xlabel('$Años de experiencia$')
ax.set_ylabel('$Pais Normalizado$')
ax.set_zlabel('$Salario$')

```
