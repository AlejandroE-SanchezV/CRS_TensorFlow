```
import pandas as pd
import numpy as np

caliHouse = pd.read_csv('https://download.mlcc.google.com/mledu-datasets/california_housing_train.csv')
caliHouse.describe()
```

```
#Clasificacióón de números
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
```

```
#Asignar nuestros datos

#Imagenes de ,los números y las etiquetas correspondientes
mnist= input_data.read_data_sets('MNIST_data', one_hot=True)

#Las imáágenes son de 28 x28 
#La capa de salida seráá un vector de 10 elementos -> 0 al 9
```

```
#Imagen -> matriz 28 x 28
#La matriz de entrada -> descompondremos la imagen en un vector 28 x28 = 784 
#none es para cuando posteriomente manejemos el arreglo
x = tf.placeholder(tf.float32,[None,784])

#Matriz de pesos  -> 784 x 10 posibles casos
P = tf.Variable(tf.zeros([784,10]))
#Variable para conocer el sesgo de nuestra red neuronal
b=tf.Variable(tf.zeros([10]))
#Ecuación para la capa de salida
y = tf.matmul(x,P)+b
#Matriz con etiquetas reales o verdaderas -> Con quien compararemos
yR = tf.placeholder(tf.float32,[None,10])
```

```
#Funcióón Coss Entropy -> medir el error que tiene el modelo

softmax =  tf.nn.softmax_cross_entropy_with_logits(labels=yR,logits=y)
```

```
#Vamos a optimizar el modelo 

costo= tf.reduce_mean(softmax)
optimizador= tf.train.GradientDescentOptimizer(0.5).minimize(costo)
```

```
#Obtendremos un arreglo de boolenaos
#Comparacióón de  nuestro vector de salida vs prediccción

prediccion = tf.equal(tf.arg_max(y,1), tf.arg_max(yR,1))

#Obtendremos la precisión el modelo
accuracy = tf.reduce_mean(tf.cast(prediccion, tf.float32))

#Vosualizacióón la efectividad en cada iteración
Produccion = tf.argmax(y,1)
init = tf.global_variables_initializer()
```

```
def avance(epoca_i,sess,last_features,last_labels):
  #Para predecir el costo de nuestra red neuronal
  costoActual = sess.run(costo,feed_dict={x: last_features, yR:last_labels})

  #Para saber la precisióón de cada iteración
  certeza = sess.run(accuracy, feed_dict={x: mnist.validation.images, 
                                          yR:mnist.validation.labels})
  print('Epoca: {:<4} - Costo: {:<8.3} Accuracy: {:<5.3}'.format(epoca_i,costoActual,certeza))
```

```
with tf.Session() as sess:
  sess.run(init)

  for epoca_i in range(1000):
    lotex,lotey = mnist.train.next_batch(100)#Asignación de valores
    sess.run(optimizador, feed_dict={x:lotex, yR:lotey})

    #NO imprimiremos todas las sesiones lo haremos de 50 
    if(epoca_i%50==0):
      avance(epoca_i, sess, lotex, lotey)
  
  print('Score: ', sess.run(accuracy, feed_dict={x: mnist.test.images, yR:mnist.test.labels}))
  print('Número que nuestra red dice que es: ', 
        sess.run(Produccion, feed_dict={x: mnist.test.images[1].reshape(1,784)})
        )
```
