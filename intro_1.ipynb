```
header = [ ('nombre','S10'), ('edad',int)] #nombre -> estring de 10 elemntos , edad -> tipo entero
datos = [('Alfonso',10),('Mario',70),('Eduardo',21),('Richie',15)]
usuarios = np.array(datos,dtype = header)

#Para ordenar
np.sort(usuarios,order= 'edad') #Indicamos que datos mandaremos y como lo queremos ordenar
#np.sort(usuarios,order= 'nombre')
```
