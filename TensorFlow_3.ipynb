```
#Le daremos al algoritmo un núúmero escrito a mano
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data #Es para clasificar los números
```

```
#Asignamos la data en el objeto
#En esta data ya hay imagenes y etiquetas 
mnist=input_data.read_data_sets("MNIST_data/", one_hot=True) 

#La imagenes tienen dimension de 28x28
#Capa de salida o vector de 10 elementos 0 al 9 ;) 
```

```
#Definición de datos y de la RED NEURONAL

#Matriz de entrada, descomponemos la imagen en vector -> 282 x 28 = 784 , none es para cuando manejemos el arreglo
x=tf.placeholder(tf.float32,[None,784]) 

#Matriz de pesos, 784 para recibir la imagen, 10 por las posible salidas
P=tf.Variable(tf.zeros([784,10])) 

#Sesgo de la red neuronal
b=tf.Variable(tf.zeros([10]))

#Capa de salida mediante la ecuación -> mostrar ecuación
y=tf.matmul(x,P)+b 

#Matriz con etiquetas reales -> La data para comparar (Es la que tiene el set de datos)
yR=tf.placeholder(tf.float32,[None,10]) 
```

```
#Funcióón Coss Entropy para medir error (Entropia) -> Esto para ajustar el modelo

#Vector mediante probabilidades 
softmax=tf.nn.softmax_cross_entropy_with_logits(labels=yR,logits=y) #Parametros yR la data real y el pronóstico

#Vamos a optimizar los costos del modelo -> Explicar el gradiente
costo=tf.reduce_mean(softmax)
optimizador=tf.train.GradientDescentOptimizer(0.5).minimize(costo)
```

```
# Obtendremos un arreglo de valores booleanos 
 #Compararemos el vector de salidas reales vs prediccióón
prediccion = tf.equal(tf.argmax(y, 1), tf.argmax(yR, 1)) #cuales estan bien y cuales no

#Veremos el accuracy del modelo
accuracy = tf.reduce_mean(tf.cast(prediccion, tf.float32))

#Vamos a ver la efectivididad del modelo en cada iteracióón
Produccion = tf.argmax(y,1)
init=tf.global_variables_initializer()
```

```
#Funcion que usaremos para ver que tan bien va a aprendiendo nuestro modelo 
def avance(epoca_i, sess, last_features, last_labels):
    #Para predecir el costo en nuesta red neuronal
    costoActual = sess.run(costo,feed_dict={x: last_features, yR: last_labels})

    #Para saber que tan preciso será en cada iteración
    Certeza = sess.run(accuracy,feed_dict={x:mnist.validation.images,yR: mnist.validation.labels})

    #print('Epoca: {} - Costo: {} Certeza: {}'.format(epoca_i,costoActual,Certeza))
    print('Epoca: {:<4} - Costo: {:<8.3} Certeza: {:<5.3}'.format(epoca_i,costoActual,Certeza))
```

```
with tf.Session() as sess:
    sess.run(init)

    for epoca_i in range(1000):#Haremos mil iteraciones
        lotex, lotey = mnist.train.next_batch(100)#Asignamos valores de la data entrenada
        sess.run(optimizador, feed_dict={x: lotex, yR: lotey})

        #No imprimiremos todas sólo de 50 en 50 
        if (epoca_i%50==0):
            avance(epoca_i, sess, lotex, lotey)#LLamamos a la funcióón y pasamos los valores



    print('Score: ',sess.run(accuracy, feed_dict={x: mnist.test.images,yR: mnist.test.labels}))

    #El reshape es para mandar sóólo una imagen
    print ('Número que nuestro algoritmo dice que es: ',sess.run(Produccion,feed_dict={x: mnist.test.images[1].reshape(1,784)}))
```
